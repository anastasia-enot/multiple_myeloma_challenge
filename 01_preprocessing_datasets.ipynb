```python
rna = pd.read_csv('data/MMRF_CoMMpass_IA9_E74GTF_Salmon_entrezID_TPM_hg19.csv')
rna
```

```python
# MMRF-CoMMpass presents some differences in formats and data respect to TCGA dataset. For example, the sample ID format in MMRF-CoMMpass is “study-patient-visit-source” (e.g.“MMRF-12341-BM” means patient 1234, first visit, from bone marrow).
```

```python
rna = rna.set_index('Unnamed: 0')
rna.index.names = ['gene']
list_cols = rna.columns.to_list()
rna.head()
```

```python
not_BM_cols = []
for col in list_cols:
    spl = col.split('_')[-1]
    if spl != 'BM':
        print(col)
        not_BM_cols.append(col)
# PB = perifpheral blood, not enough data so remove it 
```

```python
only_BM_cols = []
for col in list_cols:
    spl = col.split('_')[-1]
    if spl == 'BM':
        only_BM_cols.append(col)
print(len(only_BM_cols))
```

```python
print(only_BM_cols)
print(len(only_BM_cols))
```

```python
rna_only_BM = rna.loc[:, only_BM_cols]
rna_only_BM
```

```python
BM_1_cols = []
BM_not_1_cols = []
for col in only_BM_cols:
    spl = col.split('_')[2]
    if spl != '1':
        print(col)
        BM_not_1_cols.append(col)
    else:
        BM_1_cols.append(col)
print(len(BM_not_1_cols))
print(len(BM_1_cols))
```

```python
# are there overlapping patients in not 1
pat = []
for col in BM_not_1_cols:
    spl = col.split('_')
    patient = spl[0] + '_' + spl[1]
    pat.append(patient)
print(len(pat))
unique_pat = list(set(pat))
print(len(unique_pat))
```

```python
import collections
print([item for item, count in collections.Counter(pat).items() if count > 1])
```

```python
list_patients = ['_'.join(col.split('_')[:2]) for col in list_cols]
list_patients = list(set(list_patients))
print(len(list_patients))
print(list_patients[:10])
```

```python
dict_cols = {}
list_cols_BM = rna_only_BM.columns.to_list()
for col in list_cols_BM:
    pat = '_'.join(col.split('_')[:2])
    visit = col.split('_')[2]
    print(pat)
    print(visit)
    
    if pat not in dict_cols:
        dict_cols[pat] = [visit]
        print(dict_cols[pat])
    else:
        
        alread_existing_visits = list(dict_cols[pat])
        print(alread_existing_visits)
        new_visits = alread_existing_visits + [visit]
        dict_cols[pat] = new_visits
        print(dict_cols[pat])
        
print(dict_cols)
    

    
```

```python
# Take the last visit for patients
last_visits = {}
for pat, total_visits in dict_cols.items():
    last_visit = max(total_visits)
    last_visits[pat] = last_visit
last_visits
```

```python
# Cols that have the last visit in it
last_visit_cols_to_keep = []
for pat, visit in last_visits.items():
    col_name = pat + '_' + str(visit) + '_BM'
    last_visit_cols_to_keep.append(col_name)
    
print(last_visit_cols_to_keep)
```

```python
rna_only_last_all_genes = rna_only_BM.loc[:, last_visit_cols_to_keep]
rna_only_last_all_genes
```

```python
# take only the first visit for patients 
only_first_visit_cols = []
patient_numbers = []
for col in list_cols:
    pat = '_'.join(col.split('_')[:2])
    visit = col.split('_')[2]
    if visit == '1':
        only_first_visit_cols.append(col)
        patient_numbers.append(pat)
print(len(only_first_visit_cols))
print(len(patient_numbers))
```

```python
rna_only_first = rna_only_BM.loc[:, only_first_visit_cols]
print(rna_only_first)
```

```python
# We chose to exclude genes with a median expression of zero (i.e., not expressed in more than half the samples considered), 
#to improve the interpretability of the results, using an approach similar to that used in existing tools61 
#for differential gene expression analysis.
```

```python
med = rna_only_first.median()
```

```python
cols = rna_only_first.columns.to_list()
for i, row in rna_only_first.iterrows():
    num_zeros = 0
    for col in cols:
        if row[col] == 0:
            num_zeros += 1
    rna_only_first.loc[i, 'num_zeros'] = num_zeros/len(rna_only_first.columns.to_list())
 
```

```python
rna_only_first.head(20)      
```

```python
clean_rna_only_first = rna_only_first[rna_only_first['num_zeros'] > 0.5]
print(clean_rna_only_first.shape)
```

```python
clean_rna_only_first.to_csv('data/only_1_BM_selected_genes.csv', index=False)
```
