## Preprocess the expression data

From the original publication on the MMRF-CoMMpass study: 'MMRF-CoMMpass presents some differences in formats and data respect to TCGA dataset. For example, the sample ID format in MMRF-CoMMpass is “study-patient-visit-source” (e.g.“MMRF-12341-BM” means patient 1234, first visit, from bone marrow)'

```python
rna = rna.set_index('Unnamed: 0')
rna.index.names = ['gene']
rna = rna.rename(index = lambda x: 'gene_' + str(x))

list_cols = rna.columns.to_list()

rna.head()
```

```python
not_BM_cols = []
for col in list_cols:
    spl = col.split('_')[-1]
    if spl != 'BM':
        print(col)
        not_BM_cols.append(col)
# PB = perifpheral blood, not enough data so remove it 
```

```python
only_BM_cols = []
for col in list_cols:
    spl = col.split('_')[-1]
    if spl == 'BM':
        only_BM_cols.append(col)
print(len(only_BM_cols))
```

```python
print(only_BM_cols[:20])
print(len(only_BM_cols))
```

### From now on, we will work only on bone marrow data

```python
rna_only_BM = rna.loc[:, only_BM_cols]
rna_only_BM.head()
```

```python
BM_1_cols = []
BM_not_1_cols = []
for col in only_BM_cols:
    spl = col.split('_')[2]
    if spl != '1':
        print(col)
        BM_not_1_cols.append(col)
    else:
        BM_1_cols.append(col)
print(len(BM_not_1_cols))
print(len(BM_1_cols))
```

```python
# are there overlapping patients in not 1
pat = []
for col in BM_not_1_cols:
    spl = col.split('_')
    patient = spl[0] + '_' + spl[1]
    pat.append(patient)
print(len(pat))
unique_pat = list(set(pat))
print(len(unique_pat))
```

```python
# Create a dictionary with the number of visits per patient number
dict_cols = {}
list_cols_BM = rna_only_BM.columns.to_list()
for col in list_cols_BM:
    pat = '_'.join(col.split('_')[:2])
    visit = col.split('_')[2]
    
    if pat not in dict_cols:
        dict_cols[pat] = [visit]
    else:
        
        alread_existing_visits = list(dict_cols[pat])
        new_visits = alread_existing_visits + [visit]
        dict_cols[pat] = new_visits
        
print(len(dict_cols))
```

### Dataset BM with only the last visit

```python
# Take the last visit for patients
last_visits = {}
for pat, total_visits in dict_cols.items():
    last_visit = max(total_visits)
    last_visits[pat] = last_visit
len(last_visits)
```

```python
# Cols that have the last visit in it
last_visit_cols_to_keep = []
for pat, visit in last_visits.items():
    col_name = pat + '_' + str(visit) + '_BM'
    last_visit_cols_to_keep.append(col_name)
    
print(last_visit_cols_to_keep[:20])
```

```python
rna_only_last_all_genes = rna_only_BM.loc[:, last_visit_cols_to_keep]
rna_only_last_all_genes.head()
```

```python
# transform the dataset as to have only patient ids as rows and genes as columns

renamed_cols = []
for col in rna_only_last_all_genes.columns:
    pat = '_'.join(col.split('_', 2)[0:2])
    renamed_cols.append(pat)
    
rna_only_last_all_genes.columns = renamed_cols
rna_only_last_all_genes = rna_only_last_all_genes.T
rna.index.names = ['patient']
rna_only_last_all_genes.head()
```

```python
rna_only_last_all_genes.to_csv('data/BM_last_visit_all_genes.csv')
```

### Dataset with only the first visit for every patient

```python
# take only the first visit for patients 
only_first_visit_cols = []
patient_numbers = []
for col in only_BM_cols:
    pat = '_'.join(col.split('_')[:2])
    visit = col.split('_')[2]
    if visit == '1':
        only_first_visit_cols.append(col)
        patient_numbers.append(pat)
print(len(only_first_visit_cols))
print(len(patient_numbers))
```

```python
rna_only_first = rna_only_BM.loc[:, only_first_visit_cols]
renamed_cols = []
for col in rna_only_first.columns:
    pat = '_'.join(col.split('_', 2)[0:2])
    renamed_cols.append(pat)
    
rna_only_first.columns = renamed_cols
rna_only_first = rna_only_first.T
rna_only_first.index.names = ['patient']
rna_only_first.head()
```

```python
rna_only_first.to_csv('data/BM_first_visit_all_genes.csv')
```

## Dataset with excluded genes
### Idea taken from Schamuch et al, Nature Communications, 2020: 'exclude genes with a median expression of zero (i.e., not expressed in more than half the samples considered), to improve the interpretability of the results'.

### BM - first visit - selected genes

```python
# The BM dataset with only the first visit for each patient, and only the selected genes by the aforementioned technique. 
rna_only_first_transposed = rna_only_first.T
cols = rna_only_first_transposed.columns.to_list()
for i, row in rna_only_first_transposed.iterrows():
    num_zeros = 0
    for col in cols:
        if row[col] == 0:
            num_zeros += 1
    rna_only_first_transposed.loc[i, 'num_zeros'] = num_zeros/len(rna_only_first_transposed.columns.to_list())
rna_only_first_transposed.head()
```

```python
selected_genes_rna_only_first = rna_only_first_transposed[rna_only_first_transposed['num_zeros'] < 0.5]
selected_genes_rna_only_first.head()
```

```python
selected_genes_rna_only_first.shape
```

```python
selected_genes_rna_only_first = selected_genes_rna_only_first.T
selected_genes_rna_only_first.head()
```

```python
selected_genes_rna_only_first = selected_genes_rna_only_first.drop(['num_zeros'], axis = 1)
selected_genes_rna_only_first.to_csv('data/BM_first_visit_selected_genes.csv')
```

### BM - Last visit - selected genes

```python
# The BM dataset with only the last visit for each patient, and only the selected genes by the aforementioned technique.
rna_only_last_all_genes_transposed = rna_only_last_all_genes.T
cols = rna_only_last_all_genes_transposed.columns.to_list()
for i, row in rna_only_last_all_genes_transposed.iterrows():
    num_zeros = 0
    for col in cols:
        if row[col] == 0:
            num_zeros += 1
    rna_only_last_all_genes_transposed.loc[i, 'num_zeros'] = num_zeros/len(rna_only_last_all_genes_transposed.columns.to_list())
rna_only_last_all_genes_transposed.head()
```

```python
selected_genes_rna_only_last_all_genes = rna_only_last_all_genes_transposed[rna_only_last_all_genes_transposed['num_zeros'] < 0.5]
selected_genes_rna_only_last_all_genes.head()
```

```python
selected_genes_rna_only_last_all_genes.shape
```

```python
selected_genes_rna_only_last_all_genes = selected_genes_rna_only_last_all_genes.T
selected_genes_rna_only_last_all_genes.head()
```

## Preprocessing the clinical data

```python
display(clin_data.describe())
display(clin_data.dtypes)
```

```python
# I remove the patients with CENSORED HR_FLAG as I will not perform survival analysis
clin_data = clin_data.drop(clin_data[clin_data['HR_FLAG'] == 'CENSORED'].index)
clin_data['HR_FLAG'] = clin_data['HR_FLAG'].apply(lambda x: 1 if x == 'TRUE' else 0)
```

```python
# Remove all the columns that contain only NA values 
clin_data = clin_data.dropna(axis=1, how='all')
display(clin_data.describe())
```

```python
# Show all the columns that are in this dataset to proceed to manual analysis
clin_data.columns
```

```python
# Remove columns that are not useful or redundant
# Patient type, 'RNASeq_transLevelExpFile', 'RNASeq_geneLevelExpFile' is always the same
# 'RNASeq_transLevelExpFileSamplId', 'RNASeq_geneLevelExpFileSamplId' is the id, unique per patient
# The D_ISS columns corresponds to the stage of the disease, however some of the patients contain NaN values. Due to the lack of time, 
# I did not go into imputation techniques to handle those missing values, therefore I removed the D_ISS column

cols_to_del = ['Study', 'D_OS', 'D_OS_FLAG', 'D_PFS',
       'D_PFS_FLAG', 'D_ISS', 'PatientType', 'RNASeq_transLevelExpFile', 'RNASeq_transLevelExpFileSamplId', 
        'RNASeq_geneLevelExpFile', 'RNASeq_geneLevelExpFileSamplId',  'WES_mutationFileMutect',
        'WES_mutationFileStrelkaSNV', 'WES_mutationFileStrelkaIndel', 'WES_mutationFileStrelkaIndel']
```

```python
clean_clin_data = clin_data.drop(cols_to_del, axis=1)
display(clean_clin_data.head())
display(clean_clin_data.describe())
```

```python
# check if there are still some missing values somewhere
clean_clin_data.isna().any()
```

```python
# The column CYTO_predicted_feature_10 contains NaN, so due to the lack of time, 
# I did not go into imputation techniques to handle those missing values, therefore I removed the CYTO_predicted_feature_10 column
clean_clin_data = clean_clin_data.drop(['CYTO_predicted_feature_10'], axis=1)
display(clean_clin_data.head())
display(clean_clin_data.describe())
```

```python
# Binarize the gender column
clean_clin_data['D_Gender'] = clean_clin_data['D_Gender'].apply(lambda x: 1 if x == 'Female' else 0)
```

```python
clean_clin_data = clean_clin_data.set_index('Patient')
clean_clin_data.to_csv('data/clean_clinical_annotations.csv')
```
