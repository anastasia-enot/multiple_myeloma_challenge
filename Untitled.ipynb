```python
import pandas as pd
pd.set_option('display.max_columns', 500)
```

```python
clin_dict = pd.read_csv('data/Harmonized_Clinical_Dictionary.csv')
clin_data = pd.read_csv('data/sc3_Training_ClinAnnotations.csv')
```

```python
# Eliminate all the columns that contains only NA values 

clin_data = clin_data.dropna(axis=1, how='all')
```

```python
display(clin_data.describe())
```

```python
clin_data.columns
```

```python
# Eliminate columns that are not useful

# Patient type, 'RNASeq_transLevelExpFile', 'RNASeq_geneLevelExpFile' is always the same
# 'RNASeq_transLevelExpFileSamplId', 'RNASeq_geneLevelExpFileSamplId' is the id, unique per patient
cols_to_del = ['Study', 'D_OS', 'D_OS_FLAG', 'D_PFS',
       'D_PFS_FLAG', 'D_ISS', 'PatientType', 'RNASeq_transLevelExpFile', 'RNASeq_transLevelExpFileSamplId', 
             'RNASeq_geneLevelExpFile', 'RNASeq_geneLevelExpFileSamplId',  'WES_mutationFileMutect',
                'WES_mutationFileStrelkaSNV']
```

```python
clin_data['CYTO_predicted_feature_01'].value_counts()
```

```python
# Remove columns that have 1 or 583 different values

list_col = [col for col in clin_data.columns if (clin_data[col].nunique() == 1 or clin_data[col].nunique() == len(clin_data))]
print(list_col)
```

```python
# Remove patient from the list to delete because we will need it
list_col.remove('Patient')
print(list_col)
```

```python
# Columns to remove after sanity check
# D_ISS not useful because it's the stage, and it is probably higly correlated with death
manual_list = ['D_OS', 'D_OS_FLAG', 'D_PFS', 'D_PFS_FLAG', 'D_ISS']
```

```python
# Combine 2 lists 
to_delete = list_col + manual_list
print(to_delete)
```

```python
clean_clin_data = clin_data.drop(to_delete, axis=1)
print(clean_clin_data)
```

```python
clean_clin_data.isna().any()
```

```python
# The column CYTO_predicted_feature_10 has NA value, let's remove it
clean_clin_data = clean_clin_data.drop(['CYTO_predicted_feature_10'], axis=1)
```

```python
clean_clin_data.to_csv('data/clean_clin_data.csv', index=False)
```
