```python
not_BM_cols = []
for col in list_cols:
    spl = col.split('_')[-1]
    if spl != 'BM':
        print(col)
        not_BM_cols.append(col)
# PB = perifpheral blood, not enough data so remove it 
```

```python
only_BM_cols = list_cols.copy()
for col in not_BM_cols:
    only_BM_cols.remove(col)
print(len(only_BM_cols))
```

```python
BM_1_cols = []
BM_not_1_cols = []
for col in only_BM_cols:
    spl = col.split('_')[2]
    if spl != '1':
        print(col)
        BM_not_1_cols.append(col)
    else:
        BM_1_cols.append(col)
print(len(BM_not_1_cols))
print(len(BM_1_cols))
```
