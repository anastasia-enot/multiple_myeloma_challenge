```python
import pandas as pd
from sklearn.model_selection import train_test_split
pd. set_option('display.max_columns', 500)
```

```python
rnaseq = pd.read_csv('data/only_1_BM_all_genes.csv')
rnaseq = rnaseq.drop('num_zeros',axis=1)
new = []
for col in rnaseq.columns:
    new_c = col.replace('_1_BM', '')
    new.append(new_c)
rnaseq.columns = new
col_list = rnaseq.columns.to_list()
print(len(col_list))
col_list = list(set(col_list))
print(len(col_list))
rnaseq = rnaseq.T
display(rnaseq.head())
```

```python
print('------------')

Y = pd.read_csv('data/sc3_Training_ClinAnnotations.csv')
Y = Y.set_index('Patient')
display(Y.head())

rnaseq['HR_FLAG'] = Y['HR_FLAG']
print(len(rnaseq))
display(rnaseq.head(20))
display(rnaseq['HR_FLAG'].isnull().sum())
rnaseq = rnaseq.drop(rnaseq[rnaseq['HR_FLAG'] == 'CENSORED'].index)

print('------------')

rnaseq = rnaseq.dropna()
print(len(rnaseq))
display(rnaseq['HR_FLAG'].value_counts())

```

```python
X = rnaseq.drop(['HR_FLAG'],axis=1)

Y = rnaseq.loc[:, 'HR_FLAG']
print('----------')

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```
